```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('USvideos.csv')
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=True)
```

```python
df.info()
```

```python
df.nlargest(100, 'views')
```

```python
df.nlargest(10, ['likes','views'])
```

```python
df_top100views = df.nlargest(100, 'views');
```

```python
import seaborn
```

```python
seaborn.pairplot(df, x_vars=["views","likes","category_id"], y_vars=["views","likes","category_id"], hue="category_id",diag_kind="hist")
```

```python
seaborn.pairplot(df_top100views, x_vars=["views","likes","category_id"], y_vars=["views","likes","category_id"],diag_kind="hist")
```

```python
# Many of the relationships remain the same between the two charts
# views and likes have the same pattern the category ID comparison shows that
# there are far more limited category's on the top 100, however they seem to follow the same patterm
# Many of the lines/graphs got far thinner when looking at a smaller set, which is expected
# for the most part the main difference is the more limited number of categories
```
