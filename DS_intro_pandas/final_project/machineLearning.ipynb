```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CleanedCrimeData.csv')
```

```python
df.head()
```

```python
df2 = df
df2['Temp_High'] = df2['Temp_High'].round(-1)
df2.head(20)
```

```python
df2['rain'] = pd.to_numeric(df['rain'], errors='coerce')
```

```python
df2.info()
```

```python
import numpy as np
df2 = df2[["Temp_High", 'rain', 'OFFENSE_CODE', 'MONTH', 'DAY_OF_WEEK', 'OFFENSE_CODE_GROUP', 'DISTRICT']]
df2.replace([np.inf, np.inf], np.nan, inplace=True)
df2.dropna(inplace=True)

df2.head(10)
```

```python
df2.isnull().values.any()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df2['DAY_OF_WEEK'] = le.fit_transform(df2['DAY_OF_WEEK'] ) 
df2['OFFENSE_CODE_GROUP'] = le.fit_transform(df2['OFFENSE_CODE_GROUP'])
df2['DISTRICT'] = le.fit_transform(df2['DISTRICT'])
df2.head()
```

```python
import seaborn as sns
sns.heatmap(df2.corr(), annot=True, linewidths=.5,)
```

```python
y = df2["OFFENSE_CODE"]
X = df2.loc[:, ["Temp_High", 'rain', 'DISTRICT', 'DAY_OF_WEEK']]

print("X:", X.shape)
print("Y:", y.shape)
```

```python
from sklearn import metrics
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=2000, random_state=1)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

```python
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
from sklearn.linear_model  import LogisticRegression

logreg= LogisticRegression()

logreg.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred=logreg.predict(X_test)

print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X, y)
y_pred = knn.predict(X)
print(metrics.accuracy_score(y, y_pred))
```
