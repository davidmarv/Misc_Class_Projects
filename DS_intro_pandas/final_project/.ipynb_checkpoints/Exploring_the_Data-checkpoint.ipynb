```python
H_data = df.loc[df["Temp_High"]>=55,:]
L_data = df.loc[df["Temp_High"]<55,:]
```

```python
from pylab import rcParams
rcParams["figure.figsize"] = 30,10
order = df['OFFENSE_CODE_GROUP'].value_counts().head(10).index
seaborn.countplot(data = df, x='OFFENSE_CODE_GROUP',hue='DISTRICT', order = order);
```
